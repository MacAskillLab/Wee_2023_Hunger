```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import os
import pandas as pd
```

```python
pbs_color = '0.6'
pbs_color_r = '0.6'
ghrelin_color = 'goldenrod'
ghrelin_color_r = 'goldenrod'
```

```python
nac_color = '#a4519a'
lh_color = '#98bf98'
```

### first plot vS-NAc activity as AUC, then regression

```python
summary_NAc = np.load('AUC_NAc.npy', allow_pickle = True)
```

```python
plt.figure(figsize=(1.9,4))

ax = sns.boxplot(data =summary_NAc.T, palette = ['#a4519a'], showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.stripplot(data =np.array(summary_NAc).T,  color = '0.2')

plt.axhline(y=0, color= 'k',alpha = 0.7, linestyle='--')

plt.ylim(-600, 600)
plt.ylabel(r'AUC')
plt.xticks([0, 1, 2], ['Inv', 'Eat', 'Oth'], rotation = 45, ha="right")

plt.tight_layout()
sns.despine()
```

```python
reg_NAc = pd.read_pickle('Reg_NAc')
```

```python
plt.figure(figsize=(1.9,4))

ax = sns.boxplot(data =reg_NAc, palette = ['#a4519a'], showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.stripplot(data =reg_NAc,  color = '0.2')


plt.axhline(y=0, color= 'k',alpha = 0.7, linestyle='--')

plt.ylim(-1.5, 1.5)
plt.ylabel(r'$\beta$ Coefficient')
plt.xticks([0, 1, 2], ['Inv', 'Eat', 'Oth'], rotation = 45, ha="right")


plt.tight_layout()
sns.despine()
```

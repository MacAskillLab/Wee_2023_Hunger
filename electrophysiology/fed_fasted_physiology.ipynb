```python
from neo import io
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rc
import pandas as pd
import os
import seaborn as sns
import pickle
import scipy.signal
from scipy import stats
```

```python
from scipy import stats

def tsplot(ax, data, color, lw = 1,**kw):
    x = np.arange(data.shape[1])
    est = np.nanmean(data, axis=0)
    sd = np.nanstd(data, axis=0)/np.sqrt(data.shape[0])
    cis = (est - sd, est + sd)
    plt.fill_between(x,cis[0],cis[1],alpha=0.2, facecolor = color, **kw)
    plt.plot(x,est,**kw, c = color, linewidth = lw)
    
def mediantsplot(ax, data, color, lw = 1,**kw):
    x = np.arange(data.shape[1])
    est = np.nanmedian(data, axis=0)
    sd = np.nanstd(data, axis=0)/np.sqrt(data.shape[0])
    cis = (est - sd, est + sd)
    plt.fill_between(x,cis[0],cis[1],alpha=0.2, facecolor = color, **kw)
    plt.plot(x,est,**kw, c = color, linewidth = lw)    
```

```python
fasted_color = '#3D81AC'
sated_color = '0.6'
```

```python
NAc_data = pd.read_pickle('NAc_ephys_fed_fasted')
LH_data = pd.read_pickle('LH_ephys_fed_fasted')
```

```python
plt.figure(figsize=(1.6,2))

ax = sns.boxplot(data =NAc_data, x = 'condition', y = 'amplitude_70', palette = [sated_color, fasted_color], showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.stripplot(data =NAc_data, x = 'condition', y = 'amplitude_70',  color = '0.2')

plt.ylim(0,-600)
plt.ylabel('Amp -70 mV')
plt.xticks([0, 1], ['Fed', "Fasted"], rotation = 45, ha="right")
plt.xlabel(' ')
plt.yticks([0,-200, -400,-600], [0,200, 400,600])


plt.tight_layout()
sns.despine()
```

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import os
import seaborn as sns
import pickle
```

```python
ghrelin_color = 'goldenrod'
pbs_color = '0.6'
```

```python
NAc_amp_pbs = np.load('NAc_mini_amp_pbs.npy')
NAc_amp_ghrelin = np.load('NAc_mini_amp_ghrelin.npy')

NAc_freq_pbs = np.load('NAc_mini_freq_pbs.npy')
NAc_freq_ghrelin = np.load('NAc_mini_freq_ghrelin.npy')

LH_amp_pbs = np.load('LH_mini_amp_pbs.npy')
LH_amp_ghrelin = np.load('LH_mini_amp_ghrelin.npy')

LH_freq_pbs = np.load('LH_mini_freq_pbs.npy')
LH_freq_ghrelin = np.load('LH_mini_freq_ghrelin.npy')
```

```python
plt.figure(figsize=(1.6,4))
ax = sns.boxplot(data =[NAc_amp_pbs, NAc_amp_ghrelin], palette = [pbs_color, ghrelin_color], showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.stripplot(data =[NAc_amp_pbs, NAc_amp_ghrelin],  color = '0.2')


plt.ylim(0,-40)
plt.ylabel('Amplitude (pA)')
plt.xticks([0, 1], ['PBS', "Ghrelin"], rotation = 45)
plt.xlabel(' ')



plt.tight_layout()
sns.despine()
```

```python
plt.figure(figsize=(1.6,4))
ax = sns.boxplot(data =[NAc_freq_pbs, NAc_freq_ghrelin], palette = [pbs_color, ghrelin_color], showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.stripplot(data =[NAc_freq_pbs, NAc_freq_ghrelin],  color = '0.2')


plt.ylim(0,6)
plt.ylabel('Frequency (Hz)')
plt.xticks([0, 1], ['PBS', "Ghrelin"], rotation = 45)
plt.xlabel(' ')



plt.tight_layout()
sns.despine()
```

```python
from neo import io
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rc
import pandas as pd
import os
import seaborn as sns
import pickle
import scipy.signal
from scipy import stats
```

```python
ghrelin_color = 'goldenrod'
pbs_color = '0.6'
```

```python
NAc_data = pd.read_pickle('NAc_ephys_ghrelin_EI')
LH_data = pd.read_pickle('LH_ephys_ghrelin_EI')
```

```python
plt.figure(figsize=(1.6,2))

ax = sns.boxplot(data =NAc_data, x = 'condition', y = 'amplitude_70', palette = [pbs_color, ghrelin_color], showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.stripplot(data =NAc_data, x = 'condition', y = 'amplitude_70',  color = '0.2')


#plt.axhline(y=1, color= 'k',alpha = 0.7, linestyle='--')

plt.ylim(0,-800)
plt.ylabel('Amp -70 mV')
plt.xticks([0, 1], ['PBS', "Ghrelin"], rotation = 45, ha="right")
plt.xlabel(' ')
plt.yticks([0,-200, -400,-600,-800], [0,200, 400,600,800])


plt.tight_layout()
sns.despine()
# plt.savefig("NAc_70.pdf")
```

```python
import numpy as np
import os
import scipy
import pickle
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
pbs_color = '0.6'
ghrelin_color = 'goldenrod'
```

```python
scram_pbs, scram_ghrelin, RNAi_pbs, RNAi_ghrelin = np.load('RNAi_minis.npy', allow_pickle = True)
scram_pbs_f, scram_ghrelin_f, RNAi_pbs_f, RNAi_ghrelin_f = np.load('RNAi_minis_freq.npy', allow_pickle = True)
```

```python
plt.figure(figsize=(1.6,4))
ax = sns.boxplot(data =[scram_pbs, scram_ghrelin], palette = [pbs_color, ghrelin_color], showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.swarmplot(data =[scram_pbs, scram_ghrelin],  color = '0.2')

plt.ylim(0,-30)
plt.ylabel('Amplitude (pA)')
plt.xticks([0, 1], ['PBS', "Ghrelin"], rotation = 45)
plt.xlabel(' ')

plt.tight_layout()
sns.despine()
```

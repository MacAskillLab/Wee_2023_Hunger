```python
import numpy as np
import os
import scipy
import pickle
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
pbs_color = '0.6'
ghrelin_color = 'goldenrod'
```

```python
tra_mat = np.load('RNAi_matrices.npy', allow_pickle = True)
```

```python
# strcuture of array: [pbs_Scram, ghrelin_scram ,pbs_RNAi, ghrelin_RNAi]
```

```python
#pbs scram vs RNAi

plt.figure(figsize=(2,4))

plt.subplot(211)

sns.heatmap(data = np.mean(tra_mat[0], axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette('0.6', as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])


plt.subplot(212)
sns.heatmap(data = np.mean(tra_mat[2], axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette('0.6', as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.tight_layout()
```

```python
#ghrelin scram vs RNAi

plt.figure(figsize=(2,4))

plt.subplot(211)

sns.heatmap(data = np.mean(tra_mat[1], axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette('goldenrod', as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])


plt.subplot(212)
sns.heatmap(data = np.mean(tra_mat[3], axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette('goldenrod', as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.tight_layout()
```

```python
# pbs pinv_eat

plt.figure(figsize=(1.6,4))

data = [np.asarray(tra_mat[0])[:,0,1] , 
        np.asarray(tra_mat[2])[:,0,1]]

ax = sns.boxplot(data =data, color = pbs_color, showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.swarmplot(data =data,  color = '0.2')


plt.ylim(0,1)
plt.ylabel('p (Inv-Eat)')
plt.xticks([0, 1], ['Scram', "RNAi"], rotation = 45)
plt.xlabel(' ')



plt.tight_layout()
sns.despine()
```

```python
#ghrelin pinv_eat 
plt.figure(figsize=(1.6,4))

data = [np.asarray(tra_mat[1])[:,0,1] , 
        np.asarray(tra_mat[3])[:,0,1]]

ax = sns.boxplot(data =data, color = ghrelin_color, showfliers=False, boxprops = dict(linewidth=0, alpha = 0.8))
sns.swarmplot(data =data,  color = '0.2')

plt.ylim(0,1)
plt.ylabel('p (Inv-Eat)')
plt.xticks([0, 1], ['Scram', "RNAi"], rotation = 45)
plt.xlabel(' ')



plt.tight_layout()
sns.despine()
```

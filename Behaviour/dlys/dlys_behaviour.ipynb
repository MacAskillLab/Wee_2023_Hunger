```python
import numpy as np
import os
import scipy
import pickle
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
fed_color = '0.6'
fasted_color = '#3D81AC'

pbs_color = '0.6'
ghrelin_color = 'goldenrod'
```

```python
fasted_sal = np.load('fasted_sal.npy')
fasted_dlys = np.load('fasted_dlys.npy')
ghrelin_sal = np.load('ghrelin_sal.npy')
ghrelin_dlys = np.load('ghrelin_dlys.npy')
```

```python
plt.figure(figsize=(2,4))

plt.subplot(211)
sns.heatmap(data = np.mean(fasted_sal, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette(fasted_color, as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.subplot(212)
sns.heatmap(data = np.mean(fasted_dlys, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette(fasted_color, as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.tight_layout()
```

```python
plt.figure(figsize=(1.6,4))

data = [np.asarray(fasted_sal)[:,0,1] ,
        np.asarray(fasted_dlys)[:,0,1]]

ax = sns.boxplot(data =data, palette = [fasted_color, fasted_color], showfliers=False, boxprops = dict(linewidth=0, alpha=0.8))
sns.stripplot(data =data,  color = '0.2')

plt.ylim(0,1.1)
plt.ylabel('p (Inv-Eat)')
plt.xticks([0, 1], ['Saline', "D-lys"], rotation = 45)
plt.xlabel(' ')



plt.tight_layout()
sns.despine()
```

```python
plt.figure(figsize=(2,4))

plt.subplot(211)
sns.heatmap(data = np.mean(ghrelin_sal, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette(ghrelin_color, as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.subplot(212)
sns.heatmap(data = np.mean(ghrelin_dlys, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette(ghrelin_color, as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.tight_layout()
```

```python
import numpy as np
import os
import scipy
import pickle
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
ghrelin_color = 'goldenrod'
```

### first constant 20 Hz

```python
GFP_ON_mat = np.load('GFP_ON_mat.npy')
GFP_OFF_mat = np.load('GFP_OFF_mat.npy')
    
ChR2_ON_mat = np.load('ChR2_ON_mat.npy')
ChR2_OFF_mat = np.load('ChR2_OFF_mat.npy')
```

```python
plt.figure(figsize=(2,4))

plt.subplot(211)
sns.heatmap(data = np.mean(GFP_OFF_mat, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette('goldenrod', as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.subplot(212)
sns.heatmap(data = np.mean(GFP_ON_mat, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette('goldenrod', as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.tight_layout()
```

```python
plt.figure(figsize=(2,4))

plt.subplot(211)
sns.heatmap(data = np.mean(ChR2_OFF_mat, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette('goldenrod', as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.subplot(212)
sns.heatmap(data = np.mean(ChR2_ON_mat, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette('goldenrod', as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.tight_layout()
```

```python
plt.figure(figsize=(1.6,4))

ax = sns.boxplot(data =[np.asarray(GFP_OFF_mat)[:,0,1] ,
        np.asarray(ChR2_OFF_mat)[:,0,1]], color = ghrelin_color, showfliers=False, boxprops = dict(linewidth=0, alpha=0.8))
sns.swarmplot(data =[np.asarray(GFP_OFF_mat)[:,0,1] ,
        np.asarray(ChR2_OFF_mat)[:,0,1]],  color = '0.2')


plt.ylim(0,1.1)
plt.ylabel('p (Inv-Eat)')

plt.xticks([0, 1], ['GFP', "ChR2"], rotation = 45)
plt.xlabel(' ')



plt.tight_layout()
sns.despine()
```

```python
plt.figure(figsize=(1.6,4))
ax = sns.boxplot(data =[np.asarray(GFP_ON_mat)[:,0,1] ,np.asarray(ChR2_ON_mat)[:,0,1]], color = ghrelin_color, showfliers=False, boxprops = dict(linewidth=0, alpha=0.8))
sns.swarmplot(data =[np.asarray(GFP_ON_mat)[:,0,1] ,np.asarray(ChR2_ON_mat)[:,0,1]],  color = '0.2')

plt.ylim(0,1.1)
plt.ylabel('p (Inv-Eat)')

plt.xticks([0, 1], ['GFP', "ChR2"], rotation = 45)
plt.xlabel(' ')


plt.tight_layout()
sns.despine()
```

### now closed loop

```python
fasted_off = np.load('fasted_off.npy')
fasted_on = np.load('fasted_on.npy')
ghrelin_off = np.load('ghrelin_off.npy')
ghrelin_on = np.load('ghrelin_on.npy')
```

```python
fed_color = '0.6'
fasted_color = '#3D81AC'
```

```python
plt.figure(figsize=(2,4))

plt.subplot(211)
sns.heatmap(data = np.mean(fasted_off, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette(fasted_color, as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.subplot(212)
sns.heatmap(data = np.mean(fasted_on, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette(fasted_color, as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.tight_layout()
```

```python
plt.figure(figsize=(1.6,4))

data = [np.asarray(fasted_off)[:,0,1] ,
        np.asarray(fasted_on)[:,0,1]]

ax = sns.boxplot(data =data, palette = [fasted_color, fasted_color], showfliers=False, boxprops = dict(linewidth=0, alpha=0.8))
sns.stripplot(data =data,  color = '0.2')

plt.ylim(0,1.1)
plt.ylabel('p (Inv-Eat)')
plt.xticks([0, 1], ['Light OFF', "Light ON"], rotation = 45, ha = 'right')
plt.xlabel(' ')

plt.tight_layout()
sns.despine()
```

```python
plt.figure(figsize=(2,4))

plt.subplot(211)
sns.heatmap(data = np.mean(ghrelin_off, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette(ghrelin_color, as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.subplot(212)
sns.heatmap(data = np.mean(ghrelin_on, axis = 0), vmax = 0.75, vmin = 0, annot=True, cmap = sns.light_palette(ghrelin_color, as_cmap=True), cbar = False, linewidths = 1,
           xticklabels= ['Inv', 'Eat', 'Oth'], yticklabels= ['Inv', 'Eat', 'Oth'])

plt.tight_layout()
```

```python
plt.figure(figsize=(1.6,4))

data = [np.asarray(ghrelin_off)[:,0,1] ,
        np.asarray(ghrelin_on)[:,0,1]]

ax = sns.boxplot(data =data, palette = [ghrelin_color, ghrelin_color], showfliers=False, boxprops = dict(linewidth=0, alpha=0.8))
sns.stripplot(data =data,  color = '0.2')

plt.ylim(0,1.1)
plt.ylabel('p (Inv-Eat)')
plt.xticks([0, 1], ['Light OFF', "Light ON"], rotation = 45, ha = 'right')
plt.xlabel(' ')


plt.tight_layout()
sns.despine()

```
